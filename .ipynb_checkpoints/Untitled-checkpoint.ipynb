```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains 
import pyautogui
import time
```

```python
def getUrl(uni, year):
    url = "https://www.com.tw/cross/university_"+uni+"_"+year+".html"
    return url
```

```python
def getHtml(uni, year):
    browser = webdriver.Chrome() 
    url = getUrl(uni, year)
    print(uni, year)
    browser.get(url)
    time.sleep(5)
    links = browser.find_elements_by_partial_link_text("交叉查榜")
    # print(links)
    count = 1
    for link in links:
        actionChains = ActionChains(browser)
        actionChains.context_click(link).perform()
        print(str(count) + " right click")
        time.sleep(1)
        print(str(count) + " start pyautogui")
        pyautogui.typewrite(['k'])
        time.sleep(3)
        pyautogui.typewrite(['enter'])
        print(str(count) + " end pyautogui")
        count += 1
```

```python
uni_list = ["001", "002", "003", "004", "005","006","007","008","009",
          "011", "012", "013", "014", "015","016","017","018","019",
            "020","021", "022", "023", "025","026","027","028",
           "030","031", "032", "033", "034", "035","036","038","039",
           "040","041", "042", "043", "044", "045","046","047",
           "050","051","056","058","059",
           "060","063", "065", "079", "099", "100","101","108","109","111",
           "113","130", "132", "133", "134", "150","151","152","153","154"]
years = ["108", "107", "106", "105", "104"]
```

```python
for year in years:
    for uni in uni_list[32:]:
        getHtml(uni, year)
        
```
