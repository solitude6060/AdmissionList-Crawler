```python
import os
import pandas as pd
import numpy as np
```

```python
path = "./uni.txt"
```

```python
uni_file = open(path, "r", encoding="utf8")
```

```python
uni_data = uni_file.read()
uni_list = uni_data.split("\n")
name_sc_dict = {}
sc_list = []
name_list = []

for line in uni_list:
    #print("l", line)
    line_list = line.split(" ")
    #print(line_list)
    if line_list[-1] not in sc_list:
        sc_list.append(line_list[-1])
    if line_list[0] not in name_sc_dict:
        temp_list = []
        temp_list.append(line_list[-1])
        name_sc_dict[line_list[0]] = temp_list
    else:
        if line_list[-1] not in name_sc_dict[line_list[0]]:
            name_sc_dict[line_list[0]].append(line_list[-1])
```

```python
for name in list(name_sc_dict.keys()):
    if len(name) > 3:
        print(name)
        del name_sc_dict[name]
print(name_sc_dict)
```

```python
for name in list(name_sc_dict.keys()):
    print(name)
    name_list.append(name)
```

```python
name_dict = {}
for i, n in enumerate(name_list):
    name_dict[n] = i
```
